#### Dataframe is a main object in Pandas. It is used to represent data with rows and columns(just like excel spreadsheet).

```python
import pandas as pd
import numpy as np
```

```python
df_1 = pd.read_csv("Datasets/weather.csv")
df_1
```

```python
weather_data = {
    "date": ["2024-09-15", "2024-09-16", "2024-09-17", "2024-09-18", "2024-09-19"],
    "temperature": [25, 28, 22, 20, 27],
    "windspeed": [10, 15, 8, 5, 12],
    "event": ["Sunny", "Cloudy", "Rainy", "Snowy", "Partly Cloudy"]
}

df_2 = pd.DataFrame(weather_data)
df_2
```

```python
# Getting the shape of the dataframe
(rows1, columns1) = df_1.shape
(rows2, columns2) = df_2.shape

print("Shape of df1:",rows1,",",columns1)
print("Shape of df2:",rows2,",",columns2)
```

```python
# For printing some few rows of the data set, we use .head(), we could also pass the number of rows as argument

df_1.head()
```

```python
df = df_1.head(10)
df
```

```python
# This will provide some basic statistic data for the columns having only numerical value
df.describe()
```

```python
# For conditional selection of data, just like SQL
df[df['MinTemp'] == df['MinTemp'].max()]
```

```python
# For conditional selection of columns along with conditional data

df[['MinTemp', 'MaxTemp', 'Rainfall']][df['MaxTemp'] == df['MaxTemp'].max()]
```

```python
# Checking the current index and modifying that
print(df_2.index)

df_2.set_index('date', inplace=True) # we have used inplace, b/c this set_index function returns us a new dataframe, if we want to modify the actual dataframe then we have to set inplace = true.
```

```python
# Now we can access a particular row, more easily
df_2[['temperature', 'windspeed']].loc['2024-09-15']
```
